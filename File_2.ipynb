```python
df1 = pd.DataFrame(columns = ['Batch','Professor','Location','NumHours'])
for rowval in data['Jagannadha Rao Basa']:
#     print(rowval)
    if rowval == 0:
        continue
        
        
        
    elif "|" in rowval:
        print(rowval)
        left_part, right_part = rowval.split("|")
        if left_part.count("-") >= 2:
            newrow = {'Batch':left_part.split('-')[0],'Professor':'Jagannadha Rao Basa','Location':left_part.split('-')[1],'NumHours':left_part.split('-')[-1]}
            print(newrow)
            df1 = df1.append(newrow,ignore_index=True)
            del newrow
        elif right_part.count("-") >= 2:
            newrow = {'Batch':right_part.split('-')[0],'Professor':'Jagannadha Rao Basa','Location':right_part.split('-')[1],'NumHours':right_part.split('-')[-1]}
            print(newrow)
            df1 = df1.append(newrow,ignore_index=True)
            del newrow
        
        
        
    elif rowval.count("-") >= 2:
        newrow = {'Batch':rowval.split('-')[0],'Professor':'Jagannadha Rao Basa','Location':rowval.split('-')[1],'NumHours':rowval.split('-')[-1]}
        df1 = df1.append(newrow,ignore_index=True)
#         print(df1)
#         print('\n')
```

```python
finaldf = pd.DataFrame(columns = ['Batch','Professor','Location','NumHours'])
for prof in data.columns[1:]:
    for rowval in data[prof]:
    #     print(rowval)
        if rowval ==0:
            continue
            
        elif rowval.count("|") == 1:
            print(rowval)
            left_part, right_part = rowval.split("|")
            if left_part.count("-") >= 2:
                newrow = {'Batch':left_part.split('-')[0],'Professor':prof,'Location':left_part.split('-')[1],'NumHours':left_part.split('-')[-1]}
                print(newrow)
                df1 = df1.append(newrow,ignore_index=True)
                del newrow
            elif right_part.count("-") >= 2:
                newrow = {'Batch':right_part.split('-')[0],'Professor':prof,'Location':right_part.split('-')[1],'NumHours':right_part.split('-')[-1]}
                print(newrow)
                df1 = df1.append(newrow,ignore_index=True)
                del newrow
            
            
        if rowval.count("-") >= 2:
            newrow = {'Batch':rowval.split('-')[0],'Professor':prof,'Location':rowval.split('-')[1],'NumHours':rowval.split('-')[-1]}
            finaldf = finaldf.append(newrow,ignore_index=True)
            del newrow
    #         print(df1)
    #         print('\n')
```
