### Calculating the num of hours for Single Professor

```python
lis = []
for rowval in data['Jagannadha Rao Basa']:
    if rowval != 0:
        print(rowval.split('-')[-1])
        lis.append(rowval.split('-')[-1])
```

```python
## only adding the integer values in the list

total = 0
for element in lis:
    if isinstance(element, str):
        try:
            num = int(element) # try converting to int
        except ValueError:
            try:
                num = float(element) # try converting to float
            except ValueError:
                continue # skip this element if it cannot be converted to a number
        total += num

print("Total Lecture Hours :", total)
```

### Calculating the num of hours for all Professor

```python
final = {}
for prof in data.columns[1:]:
    total = 0
    li = []
    for rowval in data[prof]:
        if rowval != 0:
#         print(rowval.split('-')[-1])
            li.append(rowval.split('-')[-1])
    for element in li:
        if isinstance(element, str):
            try:
                num = int(element) # try converting to int
            except ValueError:
                try:
                    num = float(element) # try converting to float
                except ValueError:
                    continue # skip this element if it cannot be converted to a number
            total += num
    final[prof] = total
```

```python
final
```

```python
import csv

# Open a CSV file for writing
with open('final.csv', 'w', newline='') as csvfile:
    # Create a writer object
    writer = csv.writer(csvfile)
    
    # Write the header row
    writer.writerow(['Professor', 'NumHours'])
    
    # Write the data rows
    for key, value in final.items():
        writer.writerow([key, value])
```
